Visualize the data using Python by plotting the graphs for assignment no. 1 and 2. Consider a suitable data set. Use Scatter plot, Bar plot, Box plot, Pie chart, Line Chart.

```python
from numpy import nan
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data={
    'Name':["Harish","Ram","shital","Riya","OM","Rohan"],
    'Age':[18,19,20,21,18,18],
    'Gender':['M','M','F','F','M','M'],
    'Marks':[90,80,50,60,10,75],
    'Internship':[1.0,0.0,1.0,1.0,0.0,0.0]
}
df_f=pd.DataFrame(data)
male_count=np.sum(df_f['Gender']=='M')
female_count=np.sum(df_f['Gender']=='F')
```

# Bar Graph (Number of boys and girls )

```python
fig, ax = plt.subplots()

counts=[male_count,female_count]
bar_labels=['Male','Female']
bar_colors=['tab:red','tab:blue']
# ax.bar(df_f.Gender,counts,label=bar_labels,color=bar_colors)
ax.bar(bar_labels,counts,label=bar_labels,color=bar_colors)
ax.set_ylabel('no of students')
ax.set_title('comparison of number of male and female students')
ax.legend(title='M F')
plt.show()
```

# BOX Plot (For marks of student)

student with 10 marks is an outlier

```python
marks=df_f['Marks']
plt.boxplot(marks,vert=True,patch_artist=True,label=['marks'])
plt.title('Box Plot of Student Marks')
plt.ylabel('Marks')
plt.show()

```

```python

stats = {
    'min': np.min(marks),
    'q1': np.percentile(marks, 25),
    'median': np.median(marks),
    'q3': np.percentile(marks, 75),
    'max': np.max(marks)
}
print(stats)
```

# scatter plot

```python
df=df_f


colors = df['Gender'].map({'M': 'blue', 'F': 'red'})
plt.scatter(df['Age'], df['Marks'], c=colors, s=100, alpha=0.7)

# Add Student Names as labels
for i, txt in enumerate(df['Name']):
    plt.annotate(txt, (df['Age'][i], df['Marks'][i]), xytext=(5,5), textcoords='offset points')

plt.title('Scatter Plot: Age vs Marks (Red=Female, Blue=Male)')
plt.xlabel('Age')
plt.ylabel('Marks')
```

# Pie chart

display os market share of different operating systems

```python
import pandas as pd
import matplotlib.pyplot as plt

data_pie = {
    'OS': ['Android', 'iOS', 'Samsung', 'Others'],
    'Market_Share': [71.0, 28.3, 0.4, 0.3]
}
df_pie = pd.DataFrame(data_pie)

plt.pie(df_pie['Market_Share'], labels=df_pie['OS'], autopct='%1.1f%%', startangle=140)
plt.title('Global Mobile OS Market Share')
plt.show()
```

# Line chart

(Display How revenue changes monthly)

```python
import pandas as pd
import matplotlib.pyplot as plt

data_line = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Revenue_USD': [12000, 15000, 14500, 19000, 24000, 28000]
}
df_line = pd.DataFrame(data_line)

plt.plot(df_line['Month'], df_line['Revenue_USD'], marker='o', linestyle='-', color='b')
plt.title('Monthly Revenue Growth')
plt.xlabel('Month')
plt.ylabel('Revenue ($)')
plt.grid(True)
plt.show()
```

# Heat map

Performance of team members in a week

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


data = {
    'Monday': [45, 67, 89, 32, 55],
    'Tuesday': [56, 88, 90, 40, 62],
    'Wednesday': [70, 92, 85, 50, 78],
    'Thursday': [30, 50, 60, 20, 45],
    'Friday': [95, 85, 75, 90, 99]
}


df = pd.DataFrame(data, index=['Harish ', 'Om', 'Ram', 'Mohan', 'Sam'])

plt.figure(figsize=(10, 6))


sns.heatmap(df, annot=True, cmap='YlGnBu', linewidths=0.5)


plt.title('Weekly Team Performance Heatmap')
plt.xlabel('Days of the Week')
plt.ylabel('Team Members')

plt.show()
```

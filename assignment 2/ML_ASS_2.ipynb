# User Defined Functions

```python

def calc_mean(numbers):
  sum_num=0;
  count=0;
  for i in numbers:
    sum_num=sum_num+i
    count+=1

  return sum_num / count


def calc_min(numbers):
  minimum=numbers[0]
  for i in numbers:
    if i<minimum:
      minimum=i
  return minimum


def calc_max(numbers):
  maximum=numbers[0]
  for i in numbers:
    if i>maximum:
      maximum=i
  return maximum



def calc_median(numbers):
    sorted_nums = sorted(numbers)
    n = len(sorted_nums)
    if n % 2 == 0:
        return (sorted_nums[n//2 - 1] + sorted_nums[n//2]) / 2
    else:
        return sorted_nums[n//2]


def calc_mode(numbers):
    freq = {}
    for num in numbers:
        freq[num] = freq.get(num, 0) + 1
    max_freq = max(freq.values())
    modes = [k for k, v in freq.items() if v == max_freq]
    return modes


def calc_variance(numbers):
    mean = calc_mean(numbers)
    summation=0
    for x in numbers:
      summation=summation+(x-mean)**2
    return summation / (len(numbers) - 1)


def calc_std(numbers):
    return calc_variance(numbers)**0.5


```

```python
import pandas as pd
import numpy as np


df_f = pd.DataFrame({
    'Timestamp': pd.to_datetime(['2026-01-01 10:00', '2026-01-01 11:00', '2026-01-01 12:00', '2026-01-01 13:00']),
    'Product_ID': ['A101', 'B202', 'A101', 'C303'],
    'Category': ['Electronics', 'Home', 'Electronics', 'Garden'],
    'Sales_USD': [150.0, 200.0, 175.0, 300.0],
    'Quantity': [10, 15, 12, 8]
})


df_extra = pd.DataFrame({
    'Product_ID': ['A101', 'B202', 'C303'],
    'Supplier': ['GlobalCorp', 'LocalCo', 'GreenLeaf'],
    'Region': ['North', 'South', 'East']
})
double_df = pd.concat([df_f, df_extra], axis=0)
double_df
```

```python
pivot_summary = double_df.pivot_table(index='Category',
                                      values='Sales_USD',
                                      aggfunc='sum')

print(pivot_summary)
```

```python
pivoted_view = df_f.pivot(index='Timestamp',
                          columns='Product_ID',
                          values='Sales_USD')

print(pivoted_view)
```

```python
double_df['Sales_USD'] = double_df['Sales_USD'].interpolate()
double_df['Supplier'] = double_df['Supplier'].replace(np.nan, 'Not Available')
double_df['Sales_USD'] = double_df['Sales_USD'].replace(np.nan, 0)
double_df['Quantity'] = double_df['Quantity'].replace(np.nan, 0)
double_df['Region'] = double_df['Region'].replace(np.nan, 'West')
double_df
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(8, 6))


plt.hist(double_df['Sales_USD'], bins=6, color='skyblue', edgecolor='black')


plt.title('Frequency Distribution of Sales (USD)')
plt.xlabel('Sales Amount ($)')
plt.ylabel('Number of Transactions')


plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

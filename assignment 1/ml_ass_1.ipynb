# calculated Average of marks excluding null

```python
from numpy import nan
import pandas as pd
data={
    'Name':["Harish","Ram","shital","Riya","OM","Rohan"],
    'Age':[18,19,20,21,18,18],
    'Gender':['M','M','F','F','M','M'],
    'Marks':[90,80,50,40,nan,75],
    'Internship':[1.0,0.0,1.0,1.0,0.0,0.0]
}
df_f=pd.DataFrame(data)
df_f.dropna(how="any")['Marks'].mean()
```

```python
average_marks = df_f['Marks'].mean()
average_marks
```

```python
c=avg=0
for ele in df_f['Marks']:
  if pd.notnull(ele):
     c=c+1
     avg=avg+ele
avg=avg/c
print(avg)
```
